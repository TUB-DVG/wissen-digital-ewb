```python
for directory in os.listdir("."):
    if directory in ["TRY_2015", "TRY_2045", "change.ipynb"]:
        continue
    else:
        for sub_directory in os.listdir(directory):
            if sub_directory.endswith(".pdf"):
                os.remove(os.path.join(directory, sub_directory))
            else:
                for file in os.listdir(os.path.join(directory, sub_directory)):
                    if file.endswith(".dat"):
                        old_name = os.path.join(directory, sub_directory, file)
                        new_file = file.replace(".dat", ".txt")
                        new_file = new_file.replace("Jahr", "normal")
                        new_file = new_file.replace("Jahr", "normal")
                        new_name = os.path.join(directory, sub_directory, new_file)
                        os.rename(old_name, new_name)
                    if file.endswith(".txt"):
                        old_name = os.path.join(directory, sub_directory, file)
                        new_file = file.replace(".dat", ".txt")
                        new_file = new_file.replace("Wint", "kalt")
                        new_file = new_file.replace("Somm", "warm")
                        new_name = os.path.join(directory, sub_directory, new_file)
                        os.rename(old_name, new_name)
                    sub_names = file.split("_")
                    new_file = new_file.replace(sub_names[1], directory)
                    os.rename(os.path.join(directory, sub_directory, file), os.path.join(directory, sub_directory, new_file))

```

```python
import numpy as np
from pandas import ExcelWriter


# Entries to ignore
ignore_entries = ["SimStadt", "GeoPandas", "Szenariengenerator"]

def compare_dataframes(df1, df2, sheet_name, output_file, ignore_entries=None):
    """
    Compare two dataframes and save differences to an Excel file.
    Only compares rows where names match between dataframes.
    
    Args:
        df1: First dataframe to compare
        df2: Second dataframe to compare
        sheet_name: Name of the sheet in output Excel
        output_file: Path to save the Excel file
        ignore_entries: List of entries to ignore in comparison (default: None)
    """
    if ignore_entries is None:
        ignore_entries = ["SimStadt", "GeoPandas", "Szenariengenerator", "Scenario generator"]
        
    with ExcelWriter(output_file) as writer:
        # Ensure both dataframes have the same columns for comparison
        df2.rename(columns={"description": "shortDescription"}, inplace=True)
        common_columns = df1.columns.intersection(df2.columns)

        df1_comp = df1[common_columns]
        df2_comp = df2[common_columns]
        
        # Filter out ignored entries
        df1_comp = df1_comp[~df1_comp['name'].isin(ignore_entries)]
        df2_comp = df2_comp[~df2_comp['name'].isin(ignore_entries)]
        
        # Only keep rows where names match between dataframes
        common_names = set(df1_comp['name']) & set(df2_comp['name'])
        df1_comp = df1_comp[df1_comp['name'].isin(common_names)]
        df2_comp = df2_comp[df2_comp['name'].isin(common_names)]
        
        # Sort both dataframes by name
        df1_comp = df1_comp.sort_values('name')
        df2_comp = df2_comp.sort_values('name')
        
        try:
            # Create comparison matrix using numpy
            print({"Differences in files for sheet": sheet_name})
            comparison_values = np.equal(df1_comp.values, df2_comp.values)
            print(f"Differences found in {sheet_name} sheet: {not comparison_values.all()}")
            
            # Find and mark differences
            rows, cols = np.where(comparison_values == False)
            for item in zip(rows, cols):
                old_value = df1_comp.iloc[item[0], item[1]]
                new_value = df2_comp.iloc[item[0], item[1]]
                # Skip if both values are nan
                if pd.isna(old_value) and pd.isna(new_value):
                    continue
                name = df1_comp.iloc[item[0]]['name']
                print(f"Difference found in '{name}', column {df1_comp.columns[item[1]]}: {old_value} --> {new_value}")
                df1_comp.iloc[item[0], item[1]] = f'{old_value} --> {new_value}'
            # Save to Excel with appropriate sheet name
            df1_comp.to_excel(writer, sheet_name=sheet_name, index=False, header=True)
            print(f"Processed sheet: {sheet_name}")
            
        except Exception as e:
            print(f"Error processing sheet {sheet_name}: {str(e)}")

# For German comparison
compare_dataframes(
    df1=data_weather_german,
    df2=data_no_weather_german,
    sheet_name="German",
    output_file="./Excel_diff_german.xlsx"
)

# For English comparison
compare_dataframes(
    df1=data_weather_english,
    df2=data_no_weather_english,
    sheet_name="English",
    output_file="./Excel_diff_english.xlsx"
)
```
